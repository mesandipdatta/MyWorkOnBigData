# Bangalore-Chennai Train Analysis
This is a Python Pandas & Matplotlib project.
Objective - 
Extract the data of trains running between Bangalore and Chennai .
And plot graphs for station count for each of the trains 

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
pd.set_option('display.max_rows',10)
%pylab inline
```

```python
# read the train data 
train_data = pd.read_csv('data/traindata.csv')
```

```python
# First few rows
train_data.head()
```

```python
# Extract details of Trains starting from Bangalore station to Chennai
station_bangalore_chennai = train_data[(train_data.SourceStationCode=='SBC ') & (train_data.DestinationstationCode=='MAS ') ]
```

```python
station_bangalore_chennai
```

```python
station_bangalore_chennai - station_bangalore_chennai.set_index('trainName')
```

```python
station_bangalore_chennai
```

```python
# No of stations for each train running between Bangalore and Chennai
station_bangalore_chennai.TrainNo.value_counts()
```

```python
type(station_bangalore_chennai)
```

```python
# No of stations for each train running between Bangalore and Chennai
Station_per_train = station_bangalore_chennai.trainName.value_counts()
```

```python
Station_per_train
```

```python
type(Station_per_train)
```

```python
plt.title('No of Stations per Train')
Station_per_train.plot(label='trainName')
plt.xticks(rotation='vertical')
plt.legend()
plt.xlabel('Train Name')
plt.ylabel('No of stations')
```
